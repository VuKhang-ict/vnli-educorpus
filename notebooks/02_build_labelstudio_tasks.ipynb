```python
import json
from pathlib import Path
import pandas as pd

WORKDIR = Path("/content/vnli_labelstudio")
WORKDIR.mkdir(parents=True, exist_ok=True)

csv_url = "https://huggingface.co/datasets/ictkhangv/vnli-educorpus/resolve/main/full_demo.csv"

df = pd.read_csv(csv_url)
print(df.shape)
display(df.head())
```

```python
tasks = []

for _, row in df.iterrows():
    task = {
        "data": {
            "sample_id": str(row["id"]),
            "premise": str(row["premise"]),
            "hypothesis": str(row["hypothesis"]),
            "topic": str(row["topic"]),
            "grade_level": "" if pd.isna(row["grade_level"]) else str(row["grade_level"]),
            "subject": str(row["subject"]),
            "source_id": str(row["source_id"]),
            "difficulty": "" if pd.isna(row["difficulty"]) else str(row["difficulty"]),
            "negation_flag": "" if pd.isna(row["negation_flag"]) else str(row["negation_flag"])
        }
    }
    tasks.append(task)

print("Số task tạo được:", len(tasks))
print(json.dumps(tasks[:2], ensure_ascii=False, indent=2))
```

```python
json_path = WORKDIR / "vnli_ls_tasks_demo_v1.json"

with open(json_path, "w", encoding="utf-8") as f:
    json.dump(tasks, f, ensure_ascii=False, indent=2)

print("Đã lưu:", json_path)
print("Dung lượng file:", json_path.stat().st_size, "bytes")
```
