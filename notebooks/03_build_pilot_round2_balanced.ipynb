```python
import json
import pandas as pd
from pathlib import Path
from google.colab import files

WORKDIR = Path("/content/vnli_pilot_round2")
WORKDIR.mkdir(parents=True, exist_ok=True)

csv_url = "https://huggingface.co/datasets/ictkhangv/vnli-educorpus/resolve/main/full_demo.csv"
df = pd.read_csv(csv_url)

selected_ids = [
    "VNLI_DEMO_0003", "VNLI_DEMO_0004", "VNLI_DEMO_0007", "VNLI_DEMO_0008", "VNLI_DEMO_0010",
    "VNLI_DEMO_0011", "VNLI_DEMO_0013", "VNLI_DEMO_0017", "VNLI_DEMO_0019", "VNLI_DEMO_0020",
    "VNLI_DEMO_0021", "VNLI_DEMO_0024", "VNLI_DEMO_0026", "VNLI_DEMO_0028", "VNLI_DEMO_0030"
]

pilot2 = df[df["id"].isin(selected_ids)].copy()
pilot2 = pilot2.set_index("id").loc[selected_ids].reset_index()

pilot2[["id", "label", "topic", "difficulty", "negation_flag"]]
```

```python
tasks = []
for _, row in pilot2.iterrows():
    tasks.append({
        "data": {
            "sample_id": str(row["id"]),
            "premise": str(row["premise"]),
            "hypothesis": str(row["hypothesis"]),
            "topic": str(row["topic"]),
            "grade_level": "" if pd.isna(row["grade_level"]) else str(row["grade_level"]),
            "subject": str(row["subject"]),
            "source_id": str(row["source_id"]),
            "difficulty": "" if pd.isna(row["difficulty"]) else str(row["difficulty"]),
            "negation_flag": "" if pd.isna(row["negation_flag"]) else str(row["negation_flag"])
        }
    })

out_json = WORKDIR / "vnli_ls_tasks_pilot_round2_balanced_15.json"
with open(out_json, "w", encoding="utf-8") as f:
    json.dump(tasks, f, ensure_ascii=False, indent=2)

print("Đã tạo:", out_json)
files.download(str(out_json))
```
